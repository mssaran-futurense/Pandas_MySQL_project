Population_Update

```python
# updating the values where Population is NaN with relevant data
df_relevant_copy_fill_na.loc[ df_relevant_copy_fill_na['Population'].isna(), 'Population' ] = df_relevant_copy_fill_na['Male'] + df_relevant_copy_fill_na['Female']

# after updating the dataframe, we have only one NaN value in Population column
df_relevant_copy_fill_na[ df_relevant_copy_fill_na['Population'].isna() ]
```

```python
df_relevant_copy_fill_na.isna().sum() # as we can see the population column has only one NaN value
```

```python
# Made changes in male and female columns as well using population column
df_relevant_copy_fill_na.loc[ df_relevant_copy_fill_na['Male'].isna(), 'Male' ] = df_relevant_copy_fill_na['Population'] - df_relevant_copy_fill_na['Female']

df_relevant_copy_fill_na.loc[ df_relevant_copy_fill_na['Female'].isna(), 'Female' ] = df_relevant_copy_fill_na['Population'] - df_relevant_copy_fill_na['Male']

df_relevant_copy_fill_na.isna().sum()
```

Literate_Update

```python
df_relevant_copy_fill_na[ df_relevant_copy_fill_na['Literate'].isna() ]
```

```python
# changes made in Literate column
df_relevant_copy_fill_na.loc[ df_relevant_copy_fill_na['Literate'].isna(), 'Literate' ] = df_relevant_copy_fill_na['Literate_Male'] + df_relevant_copy_fill_na['Literate_Female']

df_relevant_copy_fill_na[ df_relevant_copy_fill_na['Literate'].isna() ]
```

```python
# updated values according to the data in Literate column

df_relevant_copy_fill_na.loc[ df_relevant_copy_fill_na['Literate_Male'].isna(), 'Literate_Male' ] = df_relevant_copy_fill_na['Literate'] - df_relevant_copy_fill_na['Literate_Female']

df_relevant_copy_fill_na.loc[ df_relevant_copy_fill_na['Literate_Female'].isna(), 'Literate_Female' ] = df_relevant_copy_fill_na['Literate'] - df_relevant_copy_fill_na['Literate_Male']

df_relevant_copy_fill_na.isna().sum()
```

Population_Update_by_Age

```python
df_relevant_copy_fill_na[ df_relevant_copy_fill_na['Young_and_Adult'].isna() ]
```

```python
df_relevant_copy_fill_na.loc[ df_relevant_copy_fill_na['Young_and_Adult'].isna(), 'Young_and_Adult' ] = df_relevant_copy_fill_na['Population'] - df_relevant_copy_fill_na['Middle_Aged'] - df_relevant_copy_fill_na['Senior_Citizen'] -df_relevant_copy_fill_na['Age_Not_Stated']
df_relevant_copy_fill_na.loc[ df_relevant_copy_fill_na['Middle_Aged'].isna(), 'Middle_Aged' ] = df_relevant_copy_fill_na['Population'] - df_relevant_copy_fill_na['Young_and_Adult'] - df_relevant_copy_fill_na['Senior_Citizen'] -df_relevant_copy_fill_na['Age_Not_Stated']
df_relevant_copy_fill_na.loc[ df_relevant_copy_fill_na['Senior_Citizen'].isna(), 'Senior_Citizen' ] = df_relevant_copy_fill_na['Population'] - df_relevant_copy_fill_na['Middle_Aged'] - df_relevant_copy_fill_na['Young_and_Adult'] -df_relevant_copy_fill_na['Age_Not_Stated']
df_relevant_copy_fill_na.loc[ df_relevant_copy_fill_na['Age_Not_Stated'].isna(), 'Age_Not_Stated' ] = df_relevant_copy_fill_na['Population'] - df_relevant_copy_fill_na['Middle_Aged'] - df_relevant_copy_fill_na['Senior_Citizen'] -df_relevant_copy_fill_na['Young_and_Adult']

df_relevant_copy_fill_na.isna().sum()
```

Households_Update

```python
df_relevant_copy_fill_na[ df_relevant_copy_fill_na['Households'].isna() ]
```

```python
# updated the data of Households
df_relevant_copy_fill_na.loc[ df_relevant_copy_fill_na['Households'].isna(), 'Households' ] = df_relevant_copy_fill_na['Households_Rural'] + df_relevant_copy_fill_na['Households_Urban']

df_relevant_copy_fill_na.isna().sum()
```

```python
df_relevant_copy_fill_na.loc[ df_relevant_copy_fill_na['Households_Rural'].isna(), 'Households_Rural' ] = df_relevant_copy_fill_na['Households'] - df_relevant_copy_fill_na['Households_Urban']

df_relevant_copy_fill_na.loc[ df_relevant_copy_fill_na['Households_Urban'].isna(), 'Households_Urban' ] = df_relevant_copy_fill_na['Households'] - df_relevant_copy_fill_na['Households_Rural']

df_relevant_copy_fill_na.isna().sum()
```

```python
# created stored functions

get_population_district = """
create function get_population_district(state_UT_name varchar(50), district_name varchar(50))
RETURNS double
DETERMINISTIC

BEGIN

RETURN(select population from census_clean where `State/UT` = state_UT_name and `District` = district_name);

END
"""


get_state_pop = """
create function get_state_pop(state_UT_name varchar(50))
RETURNS double
DETERMINISTIC

BEGIN

RETURN(select sum(population) from census_clean where `State/UT` = state_UT_name);

END
"""



senior_citizen_population = """
create function senior_citizen_population(state_UT_name varchar(50))
RETURNS double
DETERMINISTIC

BEGIN

RETURN(select sum(senior_citizen) from census_clean where `State/UT` = state_UT_name);

END
"""


get_hospital_beds = """
create function get_hospital_beds(state_UT_name varchar(50))
RETURNS double
DETERMINISTIC

BEGIN

RETURN(select HospitalBeds from all_hospitals where `State/UT` = state_UT_name);

END
"""


get_govt_hospital_beds = """
create function get_govt_hospital_beds(state_UT_name varchar(50))
RETURNS double
DETERMINISTIC

BEGIN

RETURN(select (`Rural hospitals_Beds` + `Urban hospitals_Beds`) from govt_hospitals where `States/UTs` = state_UT_name);

END
"""

beds_per_lakh = """
create function beds_per_lakh(state_UT_name varchar(50))
RETURNS double
DETERMINISTIC

BEGIN

RETURN(select (get_hospital_beds(`States/UTs`)/get_state_pop(`States/UTs`))*100000 from govt_hospitals where `States/UTs` = state_UT_name);

END
"""


govt_beds_per_lakh = """
create function govt_beds_per_lakh(state_UT_name varchar(50))
RETURNS double
DETERMINISTIC

BEGIN

RETURN(select (get_govt_hospital_beds(`States/UTs`)/get_state_pop(`States/UTs`))*100000 from govt_hospitals where `States/UTs` = state_UT_name);

END
"""


with engine.connect() as conn:
    conn.execute(sqlal.text(get_population_district))
    conn.execute(sqlal.text(get_state_pop))
    conn.execute(sqlal.text(senior_citizen_population))
    conn.execute(sqlal.text(get_hospital_beds))
    conn.execute(sqlal.text(get_govt_hospital_beds))
    conn.execute(sqlal.text(beds_per_lakh))
    conn.execute(sqlal.text(govt_beds_per_lakh))


# print(x.fetchall())

    
```

```python
create_hosp_log = """
create table hospital_log (
							`States/UTs` varchar(50),
                            `Urban/Rural` varchar(30),
                            `Action_taken` varchar(30),
                            `Date_log` date,
                            `government_private` enum('government','private')
);
"""

trigger_q_26 = """  
create trigger hosp_log_update 
AFTER
UPDATE ON govt_hospitals
FOR EACH ROW 
BEGIN
	DECLARE urban_rural varchar(50);
	DECLARE added_removed varchar(50);
    
    IF OLD.Rural_Government_Hospitals > NEW.Rural_Government_Hospitals 
    THEN
        SET urban_rural = 'Rural';
		SET added_removed = 'removed';

	ELSEIF OLD.Rural_Government_Hospitals < NEW.Rural_Government_Hospitals 
    THEN
        SET urban_rural = 'Rural';
		SET added_removed = 'added';
    
	ELSEIF OLD.Urban_Government_Hospitals > NEW.Urban_Government_Hospitals 
    THEN SET urban_rural = 'Urban';
		 SET added_removed = 'removed';

	ELSEIF OLD.Urban_Government_Hospitals < NEW.Urban_Government_Hospitals 
    THEN SET urban_rural = 'Urban';
		 SET added_removed = 'added';

    ELSE 
		SET urban_rural = 'No_changes_made';
		SET added_removed = 'No_changes_made';
        
	END IF;
    
    INSERT INTO hospital_log(
							`States/UTs`,
                            `Urban/Rural`,
                            `Action_taken`,
                            `Date_log`,
                            `government_private`
							)
	VALUES (OLD.`States/UTs`,urban_rural,added_removed,CURDATE(),'government');

END;
"""

change_1 = """  
UPDATE govt_hospitals
SET Rural_Government_Hospitals = 19823
WHERE `States/UTs` = 'All India'
"""

change_2 = """  
UPDATE govt_hospitals
SET Urban_Government_Hospitals = 20
WHERE `States/UTs` = 'Andaman & Nicobar Islands'
"""


with engine.connect() as conn:
    conn.execute(sqlal.text(create_hosp_log))
    conn.execute(sqlal.text(trigger_q_26))
    conn.execute(sqlal.text(change_1))
    conn.execute(sqlal.text(change_2))
    hosp_log = conn.execute(sqlal.text('select * from hospital_log'))

x = hosp_log.fetchall()
df_hosp_log = pd.DataFrame(x)
df_hosp_log
```

```python
create_hosp_log = """
create table hospital_count_no_log (
							`States/UTs` varchar(50),
                            `Urban/Rural` varchar(30),
                            `no_of_beds` int,
                            `Action_taken` varchar(30),
                            `Date_log` date,
                            `government_private` enum('government','private')
);
"""

trigger_q_27 = """  

create trigger hosp_count_no_log_update 
AFTER
UPDATE ON govt_hospitals
FOR EACH ROW 
BEGIN
	DECLARE urban_rural varchar(50);
	DECLARE added_removed varchar(50);
    DECLARE no_of_beds int;

    IF NEW.`Rural hospitals_Beds` > OLD.`Rural hospitals_Beds`
    THEN
        SET urban_rural = 'Rural';
        SET added_removed = 'added';
        SET no_of_beds = NEW.`Rural hospitals_Beds` - OLD.`Rural hospitals_Beds`;

    ELSEIF NEW.`Rural hospitals_Beds` < OLD.`Rural hospitals_Beds`
    THEN
        SET urban_rural = 'Rural';
        SET added_removed = 'removed';
        SET no_of_beds = OLD.`Rural hospitals_Beds` - NEW.`Rural hospitals_Beds`;

    ELSEIF NEW.`Urban hospitals_Beds` > OLD.`Urban hospitals_Beds`
    THEN
        SET urban_rural = 'Urban';
        SET added_removed = 'added';
        SET no_of_beds = NEW.`Urban hospitals_Beds` - OLD.`Urban hospitals_Beds`;

    ELSEIF NEW.`Urban hospitals_Beds` < OLD.`Urban hospitals_Beds`
    THEN
        SET urban_rural = 'Urban';
        SET added_removed = 'removed';
        SET no_of_beds = OLD.`Urban hospitals_Beds` - NEW.`Urban hospitals_Beds`;
    
    ELSE
        SET urban_rural = 'No_changes_made';
        SET added_removed = 'No_changes_made';
        SET no_of_beds = 0;
        
	END IF;
    
    INSERT INTO hospital_count_no_log(
							`States/UTs`,
                            `Urban/Rural`,
                            `no_of_beds`,
                            `Action_taken`,
                            `Date_log`,
                            `government_private`
							)
	VALUES (OLD.`States/UTs`,urban_rural,no_of_beds,added_removed,CURDATE(),'government');

END;
"""

change_1 = """  
UPDATE govt_hospitals
SET `Rural hospitals_Beds` = 580
WHERE `States/UTs` = 'Andaman & Nicobar Islands';
"""

change_2 = """  
UPDATE govt_hospitals
SET `Urban hospitals_Beds` = 70
WHERE `States/UTs` = 'Andhra Pradesh';
"""

with engine.connect() as conn:
    conn.execute(sqlal.text(create_hosp_log))
    conn.execute(sqlal.text(trigger_q_27))
    conn.execute(sqlal.text(change_1))
    conn.execute(sqlal.text(change_2))
    hosp_log = conn.execute(sqlal.text('select * from hospital_count_no_log'))

x = hosp_log.fetchall()
df_hosp_log = pd.DataFrame(x)
df_hosp_log
```
